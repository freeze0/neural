## Описание данных
### Таблица ```transactions.csv```
##### Описание
Таблица содержит историю транзакций клиентов банка за один год и три месяца.

##### Формат данных

```
customer_id,tr_datetime,mcc_code,tr_type,amount,term_id
111111,15 01:40:52,1111,1000,-5224,111111
111112,15 15:18:32,3333,2000,-100,11122233
...
```
##### Описание полей

 - ```customer_id``` — идентификатор клиента;
 - ```tr_datetime``` — день и время совершения транзакции (дни нумеруются с начала данных);
 - ```mcc_code``` — mcc-код транзакции;
 - ```tr_type``` — тип транзакции;
 - ```amount``` — сумма транзакции в условных единицах со знаком; ```+``` — начисление средств клиенту (приходная транзакция), ```-``` — списание средств (расходная транзакция);
 - ```term_id``` — идентификатор терминала;

````python
df = transactions.copy()
df
````

````python
df['tr_day'] = df['tr_datetime'].str.split().str[0]
df
````

````python
tr_day_mask = (df.groupby('tr_day').nunique()['mcc_code'] > 75)
tr_day_mask
````

````python
tr_days_uniq_mcc_greater_75 = tr_day_mask.where(tr_day_mask == True).dropna().reset_index()['tr_day']
tr_days_uniq_mcc_greater_75
````

````python
df_2 = df.set_index('tr_day').loc[tr_days_uniq_mcc_greater_75].groupby(['gender', 'mcc_code']).aggregate(amount_count=('amount', 'count'), amount_mean=('amount', 'mean'))
df_2
````

````python
df_2 = df_2.sort_values(by='amount_count', ascending=False)
df_2 = df_2.reset_index().set_index('mcc_code')
df_2 = pd.merge(df_2, tr_mcc_codes, on='mcc_code')
df_2
````
